```python
import pandas as pd
```

```python
CCTV_seoul = pd.read_csv("01. CCTV_in_Seoul.csv",encoding='utf-8')
```

```python
CCTV_seoul.rename(columns={CCTV_seoul.columns[0]:'구별'}, inplace=True) #여러개 형태면, 딕셔너리 형태로 바꿔주는것도 노하우임 { key : value}
#inplace = True는 비파괴함수를, 파괴함수로 해서 자동으로 원데이터 내용을 변경
```

```python
pop_Seoul = pd.read_excel("01. population_in_Seoul.xls", 
                           header= 2,
                           usecols = 'B, D, G, J, N',
                           encoding = 'utf-8')

#header는 x -> row '2번을' 가져온다.
#usecols 는 y -> column 'B,D<G,J' 가져옴
```

```python
pop_Seoul.sort_values(by='계', ascending=False).head()
```

```python
pop_Seoul.rename(columns={
    pop_Seoul.columns[0] : '구별',
    pop_Seoul.columns[1] : '인구수',
    pop_Seoul.columns[2] : '한국인',
    pop_Seoul.columns[3] : '외국인',
    pop_Seoul.columns[4] : '고령자',
}, inplace=True)
```

```python
pop_Seoul.tail() #테일은 아래 다섯개
```

```python
# Python means NaN : not a number, R Means NA : Not Available
# 결측치는 엑셀 파일을 변환할때 가끔 딸려온다. 그래서 데이터 결과를 찍어본 후, 결측치 처리를 한번씩 해줘야해
```

```python
#drop -> 행의 인덱스 드랍
#비파괴라 inplace로 결과반영
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
#unique = 중복제거
pop_Seoul['구별'].unique()
```

```python
pop_Seoul['구별'].isnull()
```

```python
#조건줘서 해당 열의 로우 값 검색
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
pop_Seoul.drop([26], inplace=True)
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()]
```

```python
pop_Seoul
```

```python
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
```

```python
pop_Seoul.head()
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
```

```python
pop_Seoul
```

```python
pop_Seoul.sort_values(by='고령자비율', ascending=False)
```

```python
pop_Seoul.sort_values(by='외국인비율', ascending=False)
#----끝 : 인구수별 데이터 현황  ------
```

```python
CCTV_seoul.head()
```

```python
CCTV_seoul['최근증가율'] = ((CCTV_seoul['2016년'] +
                      CCTV_seoul['2015년'] +
                      CCTV_seoul['2014년'])
                       / CCTV_seoul['2013년도 이전'] * 100 
                      )
```

```python
CCTV_seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
#열 합은 merge
#행 합은 concat 
```

```python
data_result = pd.merge(CCTV_seoul, pop_Seoul, on='구별') #열병합 
```

```python
data_result.head()
```

```python
#열삭제는 del 
#불필요한 데이터들 삭제 -> CCTV 최근증가율로 판단을 위함임
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
```

```python
data_result
```

```python
#구 별로 인덱스를 주기 위해 -> '구' -> index 변환
data_result.set_index('구별', inplace=True)
```

```python
data_result
#소계 = CCTV 계수
```

```python
#numpy 를 통해 상관관계 및 계수 분석
#np.corrcoef 상관
#두 항목간의 상관관계가 있는지 수치를 계산, 상관계수
#상관계수는 -1 ~ +1 범위의 음과 양의 관게를 가짐.
#0 : 상관없음 , 3: 약한상관, 7 강한 상관
np.corrcoef(data_result['고령자비율'], data_result['소계'])
```

```python
#t-test 는 2개 비교, 2번 이상 비교할땐 아노바 방법2번 이상 비교할땐 anova 방법
np.corrcoef(data_result['외국인비율'], data_result['소계'])
```

```python
np.corrcoef(data_result['인구수'], data_result['소계'])
```
