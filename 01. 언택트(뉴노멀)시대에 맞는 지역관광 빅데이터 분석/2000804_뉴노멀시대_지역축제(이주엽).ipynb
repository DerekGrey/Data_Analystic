```python
sido_list = {}
uniq_sidos = entrance19["시도"].unique()
uniq_sidos

sido_list2 = {}
uniq_sidos2 = entrance20["시도"].unique()
uniq_sidos2
```

```python
for sido in uniq_sidos:
    sido_sum = entrance19.loc[entrance19["시도"] == sido, "총계"].sum()
    sido_list[sido] = sido_sum
    
for sido in uniq_sidos2:
    sido_sum = entrance20.loc[entrance20["시도"] == sido, "총계"].sum()
    print(sido_sum)
    sido_list2[sido] = sido_sum
```

```python
sido_list2
```

```python
dic = pd.DataFrame.from_dict([sido_list])
dic = dic.transpose() # 행과 열 바꾸기
dic.reset_index(inplace=True) # 인덱스 리셋
dic.rename(columns = {"index": "시도", 0:"총계"}, inplace=True)

dic2 = pd.DataFrame.from_dict([sido_list2])
dic2 = dic2.transpose() # 행과 열 바꾸기
dic2.reset_index(inplace=True) # 인덱스 리셋
dic2.rename(columns = {"index": "시도", 0:"총계"}, inplace=True)
```
