```python
import pandas as pd
```

```python
from sklearn.svm import LinearSVC #학습방법 SVM(분류, 회귀) -> SVC는 Classfier(분류) -> 전류도 판단

#SVM은 추측, 회귀도 가능
```

```python
from sklearn.metrics import accuracy_score # 학습한 후, 정확도를 구하기 위해
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
iris_data = pd.read_csv('iris.csv', encoding='utf-8')
```

```python
iris_data
```

```python
#1 학습데이터 
# 주요항목 : -> 2차원 배열
learn_data = iris_data.loc[:,["SepalLength","SepalWidth","PetalLength","PetalWidth"]] #학습데이터 150개 x 4 칼럼 : 2차원 배열로 묶어줘야함

#2. 항목을 가지고 판단하는 결과 -> 1차원 배열
learn_label = iris_data.loc[:, 'Name'] #정답 = 150개
```

```python
x_train, x_test, y_train, y_test = train_test_split(learn_data, learn_label, test_size=0.2, train_size=0.8, shuffle=True) #8대2로 섞음
```

```python
x_train #학습데이터 = 문제데이터
```

```python
x_test #검증데이터(내가 검증할 데이터 문제들)
```

```python
y_train #학습 정답 데이터
```

```python
y_test #내 검증 정답 데이터
```

```python
#학습 시킬 방법(학습기, estimator,해당하는 객체 생성 
clf = LinearSVC() #객체생성 -> 선을 그릴 수 있는 (X,Y 파라미터 담을)
```

```python
clf.fit(x_train, y_train) 
#공부해 친구야..!!! 
#learn_data는 문제지(1~4), learn_label(1~4)은 정답
#머신러닝이 답을 찾으러 감
#estimator의 입력 파라메터 중 대부분은 고정값이지만, 일부는 내가 수정할 수 가 있다 => 하이퍼파라미터 
#이 하이퍼 파라미터로 정답률을 체크하고, 모델을 수정하는 것임


#LINEAR 선에서 가장 가까운 값을 찾아서, 이걸 학습시킨 후 -> 이 결과로 내 데이터를 넣어서 정확도를 측정한다.
```

```python
test_data = x_test #내가 검증할 데이터 -> 
test_label = clf.predict(test_data) #검증데이터(결과) = 답안지
```

```python
test_label #내가 검증할 데이터를 머신러닝으로 돌린 것
```

```python
#확인 결과 정확도 살펴봄 :
accuracy_score(y_test, test_label)
```
