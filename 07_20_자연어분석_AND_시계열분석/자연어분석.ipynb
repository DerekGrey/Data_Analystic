```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open('data_archive/07. alice.txt').read()
alice_mask = np.array(Image.open('data_archive/07. alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add("said") #STOPWORDS 는 해당 단어 빼겠다
```

```python
import matplotlib.pyplot as plt
import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask,
              stopwords = stopwords)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
doc_ko = kobill.open('1809890.txt').read()
#doc_ko = kobill.open('인공지능__1.csv').read()
```

```python
doc_ko
```

```python
t = Twitter()
```

```python
tokens_ko = t.nouns(doc_ko)
```

```python
import nltk
```

```python
ko = nltk.Text(tokens_ko, name='국회 의안 1809호')
```

```python
len(ko.tokens)
```

```python
len(set(ko.tokens)) # set으로 하면 중복이 없어짐
```

```python
ko.vocab() #딕셔너리 형태로 빈도수를 보여줌
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
                      '호','을','이','다','만','로','가','를', '및']

#빈도수에서 제외해줘! 불필요하거나 무의미한 단어들
```

```python
ko = [each_word for each_word in ko if each_word not in stop_words]

#foreach 문에 not in stop_words 안들어간것만 뽑아와, 
#ko foreach로 돌리는데, not in_stop_words 에 인덱스 하나씩 없는것만 뽑아오셈
```

```python
ko
```

```python
ko = nltk.Text(ko, name='국회 의안 1809호')
```

```python
ko #Text 객체로 만들어져있음
```

```python
plt.figure(figsize=(12, 6)) #빈도수에 맞춰서 그래프를 그림
ko.plot(50)
plt.show
```

```python
#Ko에 해당 단어 검색
ko.count('초등학교')
```

```python
ko.count('공무원')
```

```python
len(ko.tokens)
```

```python
len(set(ko.tokens))
#셋은 단어중복 없이 '한번만' 출력
```

```python
ko.concordance('초등학교') #연이은 -> 초등학교라는 단어와 근처에 있던 단어들을 뽑아줌!
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])

#문서의 0-600 자 사이에, 어느부분에 작성되었는지 빈도수를 보여줌
```
