## 2. 형태소 전처리 ( 의미없는 : 한글자 ) ex) 주, 등 을 없앰

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
                      '호','을','이','다','만','로','가','를', '및', '주', '등'] 

#빈도수에서 제외해줘! 불필요하거나 무의미한 단어들
```

```python
ko = [each_word for each_word in ko if each_word not in stop_words]

#foreach 문에 not in stop_words 안들어간것만 뽑아와, 
#ko foreach로 돌리는데, not in_stop_words 에 인덱스 하나씩 없는것만 뽑아오셈
```

```python
ko
```

```python
ko = nltk.Text(ko, name='인공지능기사')
```

```python
ko #Text 객체로 만들어져있음
```

```python
ko.vocab() #빈도수 분석
```

```python
plt.figure(figsize=(12, 6)) #빈도수에 맞춰서 그래프를 그림
ko.plot(50)
plt.show
```

## 3. 인공지능 관련 기사는 '주식' 과 관련된 키워드가 가장 많이 노출됨 
## 공시 > 거래 > 대비 > 지능 > 현재  > 거래량 등, AI 관련 주식 기사가 언론사에 제일 많음

```python
ko.concordance('인공') #연이은 -> 인공지능 연관
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['인공', '지능', 'A.I'])

#문서의 0-600 자 사이에, 어느부분에 작성되었는지 빈도수를 보여줌
```

```python
text = open('AI_News.txt', encoding='UTF-8').read()
alice_mask = np.array(Image.open('data_archive/it_man.png'))

stopwords = set(STOPWORDS)
stopwords.add("said") #STOPWORDS 는 해당 단어 빼겠다
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(font_path = "c:/Windows/Fonts/malgun.ttf", background_color='white', max_words=2000, mask=alice_mask,
              stopwords = stopwords)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
