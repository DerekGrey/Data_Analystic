## 2. '시계열'로 2020년 연말까지 평균 기온 예측

```python
# df 는 1990년도부터 ~ 2020년도 07.22 까지 현재 기온 데이터임
# 해당 데이터를 시계열 예측을 위한 ds 칼럼으로 바꾸어서 정제함
df = pd.DataFrame(
    {'ds' : df['날짜'], #인덱스로 넣어줌
    'y' : df['평균기온']} #무조건 y로 잡아줌
)

#날짜는 ds , 맞춰야 되는 값은 y로 만들어줘야함
```

```python
df
```

```python
m = Prophet(daily_seasonality=True)#시계열분석 예측 객체
```

```python
m.fit(df) #데이터로 시계열 예측 모델링
```

```python
future = m.make_future_dataframe(periods=163)
future

#future 라는 변수안에, 20년도 12월 31일까지 기온 예측
```

```python
forecast = m.predict(future)
forecast
#forecast 에는 시계열예측이 피처로 쓴 내용들을 담고 있음
#trends 지수, daily 수치등 데이터를 활용하여 예측을 싸라락 함!
```

```python
m.plot(forecast)

#검정색 -> 실제데이터
#파란색 -> 예측데이터 
```

## 3. 평균기온의 경우, 1990년도부터 2020년도까지 증감폭의 큰 변동이 없다.
## 그러나, 이는 '온도' 이기 때문인데, 기온이나 온도는 1도만 증감하더라도 많은 변동을 보이는 것이 사실이다.
## 또한, 계절에 따라 온도가 올라갔다 내려갔다를 반복해서 보이는 전반적인 그래프를 보이고 있음.
## 2020년도 07월 기준으로 '전년도'에 비해 그래프가 덜 상승된것으로 보아, 시계열 예측상 앞으로 날씨는 더 더워질 예정으로 예측된다.
